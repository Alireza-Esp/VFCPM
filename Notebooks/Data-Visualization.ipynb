### ðŸŸ¦ Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

### ðŸŸ¦ Getting Cleaned Data from Repository

```python
data = pd.read_csv("https://raw.githubusercontent.com/Alireza-Esp/VFCPM/refs/heads/main/Data/data-cleaned.csv")
```

```python
data
```

### ðŸŸ¦ Visualization Data

ðŸ”¹ Showing distribution of cars data across the years

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Model year" ,palette='icefire')
plt.xlabel("Model Year", fontsize=12, color='black')
plt.ylabel("Count", fontsize=12, color='black')
plt.xticks(rotation=90)
plt.show()
```

#### ðŸ”¹ Comparing manufacturers respect to the count of cars (in dataset)

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Make", order=list(data["Make"].value_counts().keys()), palette="colorblind")
plt.xlabel("Companies", fontsize=12, color='black')
plt.ylabel("Count", fontsize=12, color='black')
plt.xticks(rotation=90, size=9.5)
plt.show()
```

ðŸ”¹ Showing distribution of vehicles classes

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, y="Vehicle class", order=list(data["Vehicle class"].value_counts().keys()), palette="CMRmap")
plt.xlabel("Count", fontsize=12, color='black')
plt.ylabel("Vehicles Classes", fontsize=12, color='black')
plt.xticks(rotation=30)
plt.show()
```

#### ðŸ”¹ Showing distribution of engines size

```python
plt.figure(figsize=(10, 6))
sns.displot(data=data, x="Engine size (L)", color="teal")
plt.xlabel("Engine Size", fontsize=12)
plt.ylabel("Distribution", fontsize=12)
plt.show()
```

#### ðŸ”¹ Showing distribution of vehicles cylinders count

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Cylinders", order=list(data["Cylinders"].value_counts().keys()), palette="Set2")
plt.xlabel("Cylinders", fontsize=12, color='black')
plt.ylabel("Count", fontsize=12, color='black')
plt.show()
```

#### ðŸ”¹ Showing percentages of vehicles gearbox types

```python
plt.figure(figsize=(7,7))
plt.pie(data['Transmission'].value_counts(),
        labels=data['Transmission'].value_counts().index,
        startangle=0,
        autopct='%1.1f%%',
        colors=sns.color_palette(palette="Set3"),
        explode=[0, 0, 0, 0.3, 0.7])
plt.legend()
plt.show()
```

#### ðŸ”¹ Showing distribution of number of gears

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Number of gears", order=list(data["Number of gears"].value_counts().keys()), color="skyblue")
plt.xlabel("Number of Gears", fontsize=12, color='black')
plt.ylabel("Count", fontsize=12, color='black')
plt.show()
```

#### ðŸ”¹ Showing distribution of fuel types

```python
plt.figure(figsize=(7,7))
plt.pie(data['Fuel type'].value_counts(),
        labels=data['Fuel type'].value_counts().index,
        startangle=0,
        autopct='%1.1f%%',
        colors=sns.color_palette(palette="Set2"),
        explode=[0, 0, 0.3, 0.5, 0.9])
plt.legend(loc="upper right")
plt.show()
```

#### ðŸ”¹ Showing distribution of vehicles combined fuel consumption

```python
plt.figure(figsize=(10, 6))
sns.displot(data=data, x="Combined (L/100 km)", color="purple")
plt.xlabel("Combined Fuel Consumption (L/100km)", fontsize=12)
plt.ylabel("Distribution", fontsize=12)
plt.show()
```

#### ðŸ”¹ Showing distribution of vehicles CO2 emissions

```python
plt.figure(figsize=(10, 6))
sns.displot(data=data, y="CO2 emissions (g/km)", color="gray")
plt.xlabel("Distribution", fontsize=12)
plt.ylabel("CO2 Emissions (g/km)", fontsize=12)
plt.show()
```

#### ðŸ”¹ Showing distribution of CO2 emissions respect to Combined fuel consumption with hue Fuel type, Model year, Engine size, Cylinders, Transmission.

```python
for col in ["Fuel type", "Model year", "Engine size (L)", "Cylinders", "Transmission"]:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data, x="Combined (L/100 km)", y="CO2 emissions (g/km)", hue=col , marker="s", s=120)
    plt.xlabel("Combined Fuel Consumption (L/100 km)", fontsize=12)
    plt.ylabel("CO2 emissions (g/km)", fontsize=12)
    plt.show()
```

#### ðŸ”¹ Showing trend of average CO2 emissions and fuel consumption over time.

```python
temp_data = pd.DataFrame()
average_emission = dict()

for i in data[["Model year"]].value_counts().keys():
    average_emission.update({i[0]:sum(data.loc[data["Model year"] == i[0]]["CO2 emissions (g/km)"].values) /
                                           len(data.loc[data["Model year"] == i[0]]["CO2 emissions (g/km)"].values)})
    
temp_data.insert(0, "Model year", list(average_emission.keys()))
temp_data.insert(0, "Average CO2 emission", list(average_emission.values()))

plt.figure(figsize=(10, 6))
sns.lineplot(temp_data, x="Model year", y="Average CO2 emission")
plt.xlabel("Model year", fontsize=12)
plt.ylabel("Average CO2 emission", fontsize=12)
```

```python
temp_data = pd.DataFrame()
average_consumption = dict()

for i in data[["Model year"]].value_counts().keys():
    average_consumption.update({i[0]:sum(data.loc[data["Model year"] == i[0]]["Combined (L/100 km)"].values) /
                                           len(data.loc[data["Model year"] == i[0]]["Combined (L/100 km)"].values)})
    
temp_data.insert(0, "Model year", list(average_consumption.keys()))
temp_data.insert(0, "Average fuel consumption", list(average_consumption.values()))

plt.figure(figsize=(10, 6))
sns.lineplot(temp_data, x="Model year", y="Average fuel consumption")
plt.xlabel("Model year", fontsize=12)
plt.ylabel("Average fuel consumption", fontsize=12)
```
