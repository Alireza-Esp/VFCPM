ðŸ”· In 895 samples, "AV" is setted for "Transmission" column. should be removed.

```python
data.drop(index=list(data.loc[data["Transmission"] == "AV"].index), inplace=True)
data.reset_index(inplace=True)
data.drop(columns=["index"], inplace=True)
```

```python
data
```

ðŸ”· Separating transmission type and number of gears.

```python
Number_Of_Gears = []
NOG = ""

for index,value in zip(list(data["Transmission"].keys()), list(data["Transmission"].values)):
    
    for i in value:
        if i.isnumeric() == True:
            NOG = NOG + i
            
    Number_Of_Gears.append(int(NOG))
    data.loc[index, "Transmission"] = value[:-len(NOG)]
    NOG = ""
    
data.insert(6, "Number of gears", Number_Of_Gears)
```

ðŸ”· According to the dataset description, we change the "fuel type" column.

```python
data.loc[data["Fuel type"] == "X", "Fuel type"] = "Regular gasoline"

data.loc[data["Fuel type"] == "Z", "Fuel type"] = "Premium gasoline"

data.loc[data["Fuel type"] == "D", "Fuel type"] = "Diesel"

data.loc[data["Fuel type"] == "E", "Fuel type"] = "Ethanol (E85)"

data.loc[data["Fuel type"] == "N", "Fuel type"] = "Natural Gas"
```
